```python
# ============================================================
# STEP 1 — Load processed dataset
# ============================================================

import pandas as pd

# Path to uploaded CSV in Colab
file_path = "/content/drive/MyDrive/Asset Mapping Project/Final content/Category Formation/Processed_DFW_Assets_Master.csv"

# Load dataset
df = pd.read_csv(file_path)

# Preview dataset
print("Dataset loaded successfully.\n")

print(f"Total rows: {len(df):,}")
print(f"Total columns: {len(df.columns)}\n")

print("Columns:")
print(df.columns.tolist())

# Show first 5 rows
df.head()
```

```python
# ============================================================
# STEP 2 — Carbon Emission Reduction Analysis
# ============================================================

# ERCOT / DFW grid emission factor
# Source: EPA eGRID ERCOT subregion
# 738 lb CO2 / MWh = 0.335 kg CO2 / kWh

GRID_EMISSION_FACTOR = 0.335

# ------------------------------------------------------------
# Annual CO2 reduction per building
# ------------------------------------------------------------

df['Carbon_Reduction_kgCO2'] = (
    df['Annual_kWh']
    * GRID_EMISSION_FACTOR
).round(0)

# Convert kilograms → metric tons
df['Carbon_Reduction_tonsCO2'] = (
    df['Carbon_Reduction_kgCO2'] / 1000
).round(2)

# ------------------------------------------------------------
# Summary statistics
# ------------------------------------------------------------

print("CARBON EMISSION REDUCTION ANALYSIS COMPLETE")
print("=" * 60)

print(f"\nTotal CO2 reduction potential:")
print(f"{df['Carbon_Reduction_tonsCO2'].sum():,.0f} metric tons/year")

print(f"\nAverage CO2 reduction per building:")
print(f"{df['Carbon_Reduction_tonsCO2'].mean():,.2f} metric tons/year")

print(f"\nMaximum reduction from a single building:")
print(f"{df['Carbon_Reduction_tonsCO2'].max():,.2f} metric tons/year")

print("\nTop 10 buildings by carbon reduction:\n")

print(
    df[
        ['name',
         'building',
         'county',
         'Annual_MWh',
         'Carbon_Reduction_tonsCO2']
    ]
    .sort_values(
        'Carbon_Reduction_tonsCO2',
        ascending=False
    )
    .head(10)
    .to_string(index=False)
)
```

## Calculating cost required

```python
# ============================================================
# STEP 3 — Estimate Solar System Capacity
# ============================================================

# Physical interpretation:
# Standard solar irradiance ≈ 1 kW/m²
#
# With 20% efficient monocrystalline panels:
# 1 m² rooftop area ≈ 0.20 kW solar capacity
#
# Source:
# U.S. Department of Energy (DOE)
# Crystalline Silicon Photovoltaics Research

SYSTEM_CAPACITY_FACTOR = 0.20

# ------------------------------------------------------------
# Estimate installed solar capacity
# ------------------------------------------------------------

df['Estimated_System_kW'] = (
    df['Roof_Sq_M']
    * SYSTEM_CAPACITY_FACTOR
).round(2)

# Convert kW → MW
df['Estimated_System_MW'] = (
    df['Estimated_System_kW'] / 1000
).round(3)

# ------------------------------------------------------------
# Summary statistics
# ------------------------------------------------------------

print("SOLAR SYSTEM CAPACITY ESTIMATION COMPLETE")
print("=" * 60)

print(f"\nTotal estimated installed capacity:")
print(f"{df['Estimated_System_MW'].sum():,.2f} MW")

print(f"\nAverage system size:")
print(f"{df['Estimated_System_kW'].mean():,.1f} kW")

print(f"\nLargest estimated system:")
print(f"{df['Estimated_System_MW'].max():,.2f} MW")

print("\nTop 10 buildings by estimated capacity:\n")

print(
    df[
        ['name',
         'building',
         'county',
         'Roof_Sq_M',
         'Estimated_System_MW']
    ]
    .sort_values(
        'Estimated_System_MW',
        ascending=False
    )
    .head(10)
    .to_string(index=False)
)
```

```python
# ============================================================
# STEP 4 — Solar Installation Cost Estimation
# ============================================================

# Source:
# National Renewable Energy Laboratory (NREL)
# 2025 Commercial Rooftop PV Benchmark Cost
#
# Commercial rooftop solar cost:
# $2.0/Wdc

SOLAR_COST_PER_WATT = 2.0

# ------------------------------------------------------------
# Calculate installation cost
# ------------------------------------------------------------

# Estimated installed system cost in USD

df['Estimated_Solar_Cost_USD'] = (
    df['Estimated_System_kW']
    * 1000
    * SOLAR_COST_PER_WATT
).round(0)

# Convert to millions of dollars for readability

df['Estimated_Solar_Cost_Million_USD'] = (
    df['Estimated_Solar_Cost_USD'] / 1_000_000
).round(3)

# ------------------------------------------------------------
# Summary statistics
# ------------------------------------------------------------

print("SOLAR COST ESTIMATION COMPLETE")
print("=" * 60)

print(f"\nTotal estimated investment required:")
print(f"${df['Estimated_Solar_Cost_USD'].sum():,.0f}")

print(f"\nAverage installation cost per building:")
print(f"${df['Estimated_Solar_Cost_USD'].mean():,.0f}")

print(f"\nLargest estimated installation cost:")
print(f"${df['Estimated_Solar_Cost_USD'].max():,.0f}")

print("\nTop 10 buildings by estimated solar cost:\n")

print(
    df
    .sort_values(
        'Estimated_Solar_Cost_USD',
        ascending=False
    )[
        [
            'name',
            'building',
            'county',
            'Estimated_System_MW',
            'Estimated_Solar_Cost_Million_USD'
        ]
    ]
    .head(10)
    .to_string(index=False)
)
```
